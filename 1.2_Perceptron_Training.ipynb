```python
def Perceptron(X, y):
    weights = np.zeros(X.shape[1])  # Initialize weights
    lr = 0.01  # Learning rate
    
    for i in range(1000):
        j = np.random.randint(0, len(X))  # Random index between 0 and len(X)-1
        y_predict = step(np.dot(X[j], weights))  # Apply step function to dot product
        weights = weights + lr * (y[j] - y_predict) * X[j]  # Update weights
    
    intercept_ = weights[0]  # The first element is the intercept
    coef_ = weights[1:]  # The remaining elements are the coefficients
    
    return intercept_, coef_

# Example call to the Perceptron function
intercept_, coef_ = Perceptron(X, y)
```
